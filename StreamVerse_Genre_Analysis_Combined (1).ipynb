# StreamVerse Content Strategy Analysis
## Objective 1 — The Genre Puzzle

This notebook analyzes TV show genres to understand audience reception and ratings.
It merges three datasets and visualizes how genres perform.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
```

## Load Datasets

```python
metadata = pd.read_csv('shows_metadata.csv')
info = pd.read_csv('shows_info.csv')
ratings = pd.read_csv('shows_ratings.csv')

print(metadata.shape)
print(info.shape)
print(ratings.shape)
```

## Clean show names for merging

```python
metadata['show_name'] = metadata['show_name'].str.lower().str.strip()
info['show_name'] = info['show_name'].str.lower().str.strip()
ratings['show_name'] = ratings['show_name'].str.lower().str.strip()
```

## Merge datasets

```python
df = metadata.merge(info, on='show_name')
df = df.merge(ratings, on='show_name')

df.head()
```

## Genre Distribution
This chart shows the number of shows available in each genre.

```python
genre_counts = df['genres'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=genre_counts.values, y=genre_counts.index)
plt.title('Top Genres by Number of Shows')
plt.xlabel('Number of Shows')
plt.ylabel('Genre')
plt.show()
```

## Average Rating by Genre

```python
genre_rating = df.groupby('genres')['rating'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=genre_rating.values, y=genre_rating.index)
plt.title('Top Genres by Average Rating')
plt.xlabel('Average Rating')
plt.ylabel('Genre')
plt.show()
```

## Distribution Analysis (Charts Like the Example)
Below are multiple distribution charts showing how key variables are spread across the dataset.

```python
columns = ['rating','votes','popularity','release_year','number_of_seasons','status']
colors = ['blue','red','green','gold','purple','orange']

plt.figure(figsize=(16,10))

for i,col in enumerate(columns):
    plt.subplot(2,3,i+1)
    sns.histplot(df[col], bins=30, kde=True, color=colors[i])
    plt.title(f'{col} Distribution')
    plt.xlabel(col)
    plt.ylabel('Frequency')

plt.suptitle('TV Show Dataset Distributions', fontsize=16)
plt.tight_layout()
plt.show()
```

## Insights

- Genres with higher ratings indicate better audience reception.
- Distribution charts help understand how ratings, votes and popularity vary.
- These insights help StreamVerse decide which genres are worth investing in.
